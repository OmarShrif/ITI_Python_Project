```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import math
from datetime import datetime

def no(x):
    pass

img=np.zeros((500,500,3),np.uint8)
cv2.namedWindow('Clock')
font=cv2.FONT_HERSHEY_SIMPLEX
while True:
    img=np.zeros((500,500,3),np.uint8)
    now = datetime.now()
    current_date = now.strftime("%Y-%m-%d")
    current_time = now.strftime("%H:%M:%S")
    cv2.putText(img,"Data : %s" %current_date,(20,50),font,1,(200,200,0),2)
    cv2.putText(img,"Time : %s" %current_time,(20,90),font,1,(200,200,0),2)
    cv2.circle(img,(250,310),150,(0,0,255),2)
#Clock arranging
    for i in range(12):
        c1=int(250+150*(math.cos(math.radians(360-i*30))))
        c2=int(310+150*(math.sin(math.radians(360-i*30))))
        if 1<=i<=2:
            d1=int(250-160*(math.sin(math.radians(360-i*30))))
            d2=int(320-170*(math.cos(math.radians(360-i*30))))
        elif 3<=i<=4:
            d1=int(250-160*(math.sin(math.radians(360-i*30))))
            d2=int(320-160*(math.cos(math.radians(360-i*30))))
        elif i==5:
            d1=int(240-180*(math.sin(math.radians(360-i*30))))
            d2=int(310-185*(math.cos(math.radians(360-i*30))))
        elif i==6:
            d1=int(240-200*(math.sin(math.radians(360-i*30))))
            d2=int(310-185*(math.cos(math.radians(360-i*30))))
        elif i==7:
            d1=int(240-180*(math.sin(math.radians(360-i*30))))
            d2=int(310-185*(math.cos(math.radians(360-i*30))))
        elif i==8:
            d1=int(250-190*(math.sin(math.radians(360-i*30))))
            d2=int(310-190*(math.cos(math.radians(360-i*30))))
        elif i==9:
            d1=int(250-180*(math.sin(math.radians(360-i*30))))
            d2=int(310-160*(math.cos(math.radians(360-i*30))))
        elif i==10:
            d1=int(250-200*(math.sin(math.radians(360-i*30))))
            d2=int(310-160*(math.cos(math.radians(360-i*30))))
        elif i==11:
            d1=int(250-210*(math.sin(math.radians(360-i*30))))
            d2=int(310-165*(math.cos(math.radians(360-i*30))))
        elif i==0:
            d1=int(230-160*(math.sin(math.radians(360-i*30))))
            d2=int(310-160*(math.cos(math.radians(360-i*30))))

        if i==0:
            i=12
        cv2.putText(img,"%s" %i,(d1,d2),font,1,(200,200,0),2)
        cv2.circle(img,(c1,c2),2,(0,255,255),2)
    for i in range(60):
        p1=int(250+150*(math.cos(math.radians(360-i*6))))
        p2=int(310+150*(math.sin(math.radians(360-i*6))))
        cv2.circle(img,(p1,p2),1,(50,50,50),2)

#Calculations
    sec= int(now.strftime("%S"))
    min= int(now.strftime("%M"))
    hour= int(now.strftime("%H"))
    if hour>=12:
        hour= int((int(now.strftime("%H")))-12)
    x0=int(250-70*(math.sin(math.radians(360-hour*30))))
    y0=int(310-70*(math.cos(math.radians(360-hour*30))))
    x1=int(250-90*(math.sin(math.radians(360-min*6))))
    y1=int(310-90*(math.cos(math.radians(360-min*6))))
    x2=int(250-110*(math.sin(math.radians(360-sec*6))))
    y2=int(310-110*(math.cos(math.radians(360-sec*6))))
    cv2.line(img,(250,310),(x0,y0),(150,150,0),2) #Hour blue
    cv2.line(img,(250,310),(x1,y1),(0,200,0),2) #Min light green
    cv2.line(img,(250,310),(x2,y2),(48,70,150),2) #Sec broun
    cv2.imshow('Clock',img)
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```
